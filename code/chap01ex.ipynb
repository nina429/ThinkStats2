Count the number of times each value occurs.

```python
# 以下开始使用点标记法进行一些数据获取，怀孕结果的整数代码 
# value_counts 方法，用于计算每个值出现的次数
# sort_index 方法，将 series 对象按索引排序，是结果按序显示
preg.outcome.value_counts().sort_index()
```

```python
# 不排序返回结果，左边是新生儿磅数
preg.birthwgt_lb.value_counts(sort=False)
```

```python
# value_counts 方法，用于计算每个值出现的次数
# 进行了排序
preg.birthwgt_lb.value_counts().sort_index()
```
